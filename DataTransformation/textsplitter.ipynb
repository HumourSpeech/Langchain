##### Text Splitting from Documents

```python
from langchain_community.document_loaders import ArxivLoader
docs = ArxivLoader(query="1706.03762", load_max_docs=4).load()
```

##### How to recursively split text by characters

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=500,
    chunk_overlap=50,
)

final_docs = text_splitter.split_documents(docs)
```

```python
final_docs[0].page_content, final_docs[1].page_content
```

##### Converting text file object into document

```python
from langchain_community.document_loaders import TextLoader
loader = TextLoader('speech.txt')
text_docs = loader.load()
text_docs
```

```python
with open('speech.txt') as file:
    speech = file.read()

text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=50,
    chunk_overlap=10,
)
text = text_splitter.create_documents([speech])
text[0].page_content, text[1].page_content
```

##### Character text splitter(uses seperator to split into chunks and '\n\n' is default seperator)

```python
from langchain_text_splitters import CharacterTextSplitter
text_splitter = CharacterTextSplitter(
    separator="\n\n",
    chunk_size=100,
    chunk_overlap=20,
)
text = text_splitter.create_documents([speech])
```
