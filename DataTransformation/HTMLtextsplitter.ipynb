##### Creating documents with HTMLHeaderTextsplitter from html content or direct url

```python
from langchain_text_splitters import HTMLHeaderTextSplitter

html_string = """
<!DOCTYPE html>
<html>
<head>
    <title>Test HTML</title>
</head>
<body>
    <div>
        <h1>Header 1</h1>
        <p>This is a paragraph.</p>
        <h2>Header 2</h2>
        <br>
        <p>This is another paragraph.</p>
        <div>Additional content</div>
    </div>
</body>
</html>
"""

headers_to_split_on = [("h1","Header 1"), ("h2","Header 2")]

html_splitter = HTMLHeaderTextSplitter(headers_to_split_on)
html_header_splits = html_splitter.split_text(html_string)

html_header_splits
```

```python
url = "https://techcrunch.com/2025/11/04/sora-is-now-available-on-android-in-the-us-canada-and-other-regions/"
headers_to_split_on = [("h1","Header 1"), ("h2","Header 2"), ("h3","Header 3")]

html_splitter_from_url = HTMLHeaderTextSplitter(headers_to_split_on)
html_header_splits_from_url = html_splitter_from_url.split_text_from_url(url)
html_header_splits_from_url[6].page_content # Displaying first 500 characters of the content
```
