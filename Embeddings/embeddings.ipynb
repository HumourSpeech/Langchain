##### Embedding techniques (Converting text chunks into vectors before storing in vector database.)

```python
from langchain_openai import OpenAIEmbeddings
embeddings = OpenAIEmbeddings(model="text-embedding-3-large")
embeddings
```

```python
text = "LangChain is a framework for developing applications powered by language models. It can be used for chatbots, Generative Question-Answering (GQA), summarization, and much more."
query_result = embeddings.embed_query(text)
query_result
```

```python
len(query_result)
```

```python
embeddings_1024 = OpenAIEmbeddings(model="text-embedding-3-1024", dimensions = 1024)
```

```python
embeddings_1024
```

```python
## Loading Text from source
from langchain_community.document_loaders import TextLoader
loader = TextLoader('speech.txt')
text_docs = loader.load()
text_docs
```

```python
## Transforming data into smaller chunks(documents)
from langchain_text_splitters import RecursiveCharacterTextSplitter
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=500,
    chunk_overlap=50,
)

final_docs = text_splitter.split_documents(text_docs)
final_docs
```

```python
## Vector Embedding and Vector StoreDB
from lanchain_community.vectorstores import Chroma
DB = Chroma.from_documents(final_docs, embeddings_1024)
DB
```

```python
## Retrieving results from vector store based on similarity search technique
query = "It will be all the easier for us to conduct purselves as belligerets"
retrieved_results = DB.similarity_search(query)
print(retrieved_results)
```
