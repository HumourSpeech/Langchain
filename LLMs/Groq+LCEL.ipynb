<h5>Understanding Langchain Expression Language with Groq APIs</h5>

```python
import os
from dotenv import load_dotenv
load_dotenv()
```

```python
GROQ_API_KEY = os.getenv("GROQ_API_KEY")

## Langsmith Tracking
os.environ["LANGCHAIN_API_KEY"] = os.getenv("LANGCHAIN_API_KEY")
os.environ["LANGCHAIN_TRACING_V2"] = "true"
os.environ["LANGCHAIN_PROJECT"] = os.getenv("LANGCHAIN_PROJECT")
```

```python
from langchain_groq import ChatGroq
model = ChatGroq(model="openai/gpt-oss-120b", groq_api_key=GROQ_API_KEY)
model
```

```python
# model.generate_prompt(["Hello, my name is Nitin Mishra."])
```

```python
from langchain_core.messages import HumanMessage, SystemMessage
messages = [
    SystemMessage(content="Translate the folowing from ENglish to French."),
    HumanMessage(content="Hello, my name is Nitin Mishra.")
]
```

```python
result = model.invoke(messages)
```

```python
from langchain_core.output_parsers import StrOutputParser
parser = StrOutputParser()
parser.invoke(result)
```

```python
## Using LCEL  - for chaining the calls together
chain = model | parser
chain.invoke(messages)
```

```python
## Prompt Templates
from langchain_core.prompts import ChatPromptTemplate

generic_template = "Translate the following into {langauge}:"

prompt = ChatPromptTemplate.from_messages(
    [("system", generic_template), ("user", "{text}")]
)
```

```python
result = prompt.invoke({"langauge": "French", "text": "Hello, my name is Nitin Mishra."})
```

```python
result.to_messages()
```

```python
chain = prompt | model | parser
chain.invoke({"langauge": "French", "text": "Hello, my name is Nitin Mishra."})
```
